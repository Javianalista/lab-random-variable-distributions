```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
data
```

```python
data.isna().sum()
```

```python
numerical = data._get_numeric_data()
numerical
```

```python
def plots(row):
    for column in row.columns:
        sns.displot(row[column])
        plt.show()  
    return
```

```python
plots(numerical)
```

```python
# Most of the variables fit the exponential distribution.
```

```python
transformer = MinMaxScaler().fit(numerical)
x_normalized = transformer.transform(numerical)
print(x_normalized.shape)
numnor = pd.DataFrame(x_normalized, columns=numerical.columns)
numnor
```

```python
plots(numnor)
```

```python
Standardtransformer = StandardScaler().fit(numerical)
x_standardized = Standardtransformer.transform(numerical)
print(x_standardized.shape)
x_standardized = pd.DataFrame(x_standardized,columns=numerical.columns)
x_standardized.head()
```

```python
plots(x_standardized)
```

```python
# For this two methods it's the same distribution for all but in a different scale.
```

```python
def log_transfom_clean1(x):
    for column in numerical.columns:
        x[column] = x[column].astype(float)
        for i in range(len(x[column])):
            log = np.log10(x[column][i])
            if np.isfinite(log):
                x[column][i] = log
            else:
                x[column][i] =  0
    return x
```

```python
log = log_transfom_clean1(numerical)
log
```

```python
plots(log)
```

```python
# These new distributions are more similar(only in some cases) to a normal distribution than the others, which were more like a exponential.
```

```python
# Trying another log_transform
```

```python
def log_transfom_clean2(x):
    for column in numerical.columns:
        x[column] = x[column].astype(float)
        for i in range(len(x[column])):
            log = np.log10(x[column][i])
            if np.isfinite(log):
                x[column][i] = log
            else:
                x[column][i] =  np.NAN
        mean = x[column].mean()
        x[column] = x[column].replace(np.nan,mean)
    return x
```

```python
log2 = log_transfom_clean2(numerical)
log2
```

```python
plots(log2)
```

```python
# Definitely they are more similar to a normal distribution that the non-transformed data, but they are very far to be a normal distribution equally.
```
